```python
import csv

import matplotlib.pyplot as plt 
f= open("seoul.csv",encoding="cp949")
data=csv.reader(f)

header = next(data)

result=[]

for row in data:
    if row[4] !="":
        if row[0].split("-")[1]=="12":
          result.append(float(row[4]))
      

f.close()
#print(len(result))

plt.figure(figsize=(10,2))
plt.plot(result,"r")
plt.xlabel("Day(December)")
plt.ylabel("The highest temperature")
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt 

f= open("seoul.csv",encoding="cp949")
data=csv.reader(f)

header = next(data)

result=[]

for row in data:
    if row[4] !="":
        if (row[0].split("-")[1]=="12") and (row[0].split("-")[2]=="25"):
          result.append(float(row[4]))
      

f.close()
#print(len(result))

plt.figure(figsize=(10,2))
plt.plot(result,"r")
plt.xlabel("Day")
plt.ylabel("The highest temperature")
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt 

f= open("seoul.csv",encoding="cp949")
data=csv.reader(f)

header = next(data)

mean=[]
low=[]
high=[]

for row in data:
    if (row[2] !="")and (row[3]!="")and (row[4]!=""):
        if (row[0].split("-")[1]=="12") and (row[0].split("-")[2]=="25"):
          mean.append(float(row[2]))
          low.append(float(row[3]))
          high.append(float(row[4]))
      

f.close()
#print(len(result))

plt.figure(figsize=(10,2))
plt.plot(mean,"k")
plt.plot(low,"b")
plt.plot(high,"r")
plt.xlabel("Day")
plt.ylabel("The highest temperature")
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt
f=open("seoul.csv", encoding="cp949") 
data=csv.reader(f)
header=next(data)

month=[[],[],[],[],[],[],[],[],[],[],[],[]]

for row in data:
    if row[4] !="":
        month[int(row[0].split("-")[1])-1].append(float(row[4]))
plt.boxplot(month)
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt
f=open("seoul.csv", encoding="cp949") 
data=csv.reader(f)
header=next(data)

day=[]
for i in range(31):
    day.append([])

for row in data:
    if row[4] !="":
        if row[0].split("-")[1]=="01":
            day[int(row[0].split("-")[2])-1].append(float(row[4]))
plt.style.use("ggplot")
plt.figure(figsize=(10,5))
plt.boxplot(day,showfliers=False)
plt.show()
```
