```python
import numpy as np
import matplotlib.pyplot as plt

x= np.random.randint(-100,100,1000)
y= np.random.randint(-100,100,1000)

mask1=(abs(x)>50)
mask2=(abs(y)>50)
x= x[mask1+mask2]
y=y[mask1+mask2]

plt.scatter(x,y,c=x,cmap='jet',alpha=0.7)
plt.colorbar()
plt.show()
```
